# Vectors and Matrices 

```python
import numpy as np
from qiskit.visualization import array_to_latex

# Define vectors to represent quantum states
ket0 = np.array([[1], [0]])
ket1 = np.array([[0], [1]])

# Define to represent operations
M1 = np.array([[1, 1], [0, 1]])
M2 = np.array([[0, 0], [1, 1]]) 
display(array_to_latex(M1))
display(array_to_latex(M2))
```

```python
# Matrix multiplicaiton
stateA = np.matmul(M1, ket0)
stateB = np.matmul(M1, ket1)

M3 = np.matmul(M1, M2)
M4 = np.matmul(M2, M1)

display(array_to_latex(stateA))
display(array_to_latex(stateB))
display(array_to_latex(M3))
display(array_to_latex(M4))
```
