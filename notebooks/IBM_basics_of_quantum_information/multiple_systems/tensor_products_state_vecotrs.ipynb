```python
from qiskit.quantum_info import Statevector, Operator
from numpy import sqrt
from IPython.display import display, Latex, Math
```

### ðŸ”µ Example 1

Use the `tensor` method to compute the vector 
$$\ket{\psi_1} = \ket{0} \otimes \ket{1}.$$

```python
zero = Statevector([1, 0])
one = Statevector([0, 1])
psi_1 = zero.tensor(one)

display(Math(r'\psi_1=\ket{0}\otimes\ket{1}=' + psi_1.draw("latex_source")))
```

#### More common state vectors

```python
# X Basis
plus = Statevector.from_label("+")
minus = Statevector.from_label("-")

# Y Basis
plus_i = Statevector.from_label("r")
minus_i = Statevector.from_label("l")
```

```python
print('X Basis:')
display(Math(r'\ket{+}=' + plus.draw("latex_source")))
display(Math(r'\ket{-}=' + minus.draw("latex_source")))
```

```python
print('Y Basis:')
display(Math(r'\ket{i}=' + plus_i.draw("latex_source")))
display(Math(r'\ket{-i}=' + minus_i.draw("latex_source")))
```

### ðŸ”µ Example 2

Use the `^` method to compute the vector
$$\psi_2 = \ket{+} \otimes \ket{-i}.$$

```python
psi_2 = plus ^ minus_i
display(Math(r'\ket{\psi_2}=' + psi_2.draw("latex_source")))
```
