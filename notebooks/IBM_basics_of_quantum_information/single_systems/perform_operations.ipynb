### Operations with `Operator` and `Statevector`

```python
from qiskit.quantum_info import Operator
from numpy import sqrt
from IPython.display import display, Latex

Y = Operator([[0, -1.0j], [1.0j, 0]])
H = Operator([[1 / sqrt(2), 1 / sqrt(2)], [1 / sqrt(2), -1 / sqrt(2)]])
S = Operator([[1, 0], [0, 1.0j]])
T = Operator([[1, 0], [0, (1 + 1.0j) / sqrt(2)]])

display(Latex(r'$Y=$' + Y.draw('latex_source')))
display(Latex(r'$H=$' + H.draw('latex_source')))
display(Latex(r'$S=$' + S.draw('latex_source')))
display(Latex(r'$T=$' + T.draw('latex_source')))
```

Apply unitary operations using the `evolve` method

```python
from qiskit.quantum_info import Statevector

state0 = Statevector([1, 0])    # initialize qubit

state1 = state0.evolve(H)       # apply H-gate to initial state
state2 = state1.evolve(T)
state3 = state2.evolve(H)
state4 = state3.evolve(S)
state5 = state4.evolve(Y)

print('State 0:')
display(state0.draw("latex"))
print('State 1:')
display(state1.draw("latex"))
print('State 2:')
display(state2.draw("latex"))
print('State 3:')
display(state3.draw("latex"))
print('State 4:')
display(state4.draw("latex"))
print('State 5:')
display(state5.draw("latex"))
```
