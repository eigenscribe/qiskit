### Vectors and Matrices

```python
import numpy as np
from qiskit.visualization import array_to_latex

# Define vectors to represent quantum states
ket0 = np.array([[1], [0]])
ket1 = np.array([[0], [1]])

# Define to represent operations
M1 = np.array([[1, 1], [0, 1]])
M2 = np.array([[0, 0], [1, 1]]) 
display(array_to_latex(M1))
display(array_to_latex(M2))
```

```python
# Matrix-vector multiplicaiton
ketA = np.matmul(M1, ket0)
ketB = np.matmul(M1, ket1)

display(array_to_latex(ketA))
display(array_to_latex(ketB))
```

### State Vectors

```python
from qiskit.quantum_info import Statevector
from numpy import sqrt

u = Statevector([1 / sqrt(2), 1 / sqrt(2)])
v = Statevector([(1 + 2.0j) / 3, -2/3])
w = Statevector([1 / 3, 2 / 3])
```

```python
display(u.draw("text"))
display(u.draw("latex"))
print(u.draw("latex_source"))
```

```python
display(u.is_valid())
display(v.is_valid())
display(w.is_valid())
```
