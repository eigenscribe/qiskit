```python
from utils.qubit_cond_probability import *

from qiskit.quantum_info import Statevector, Operator, partial_trace
from qiskit import QuantumCircuit
import numpy as np
from numpy import sqrt
from IPython.display import display, Latex, Math

from qiskit import __version__
print(__version__)
```

```python
qc = QuantumCircuit(3)

# Create Bell pair between qubit 0 (Clancy) and qubit 1 (Referee)
qc.h(2)
qc.cx(2, 1)

# Custom superposition on qubit 0 ()
# Example: |psi> = sqrt(1/3)|0> + \sqrt(2/3)|1>
theta = 2 * np.arccos(np.sqrt(1/3))
qc.ry(theta, 0)

display(qc.draw('mpl'))

# Convert to statevector
sv = Statevector.from_instruction(qc)
print("Initial statevector for\n")
display(sv.draw("latex"))
```

```python
final_st, cond_prob = analyze_circuit(qc, target_qubit=2, target_bit=1, init_state="010")
display(final_st.draw("latex"))
print(cond_prob)
```
